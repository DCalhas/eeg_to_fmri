```python
import matplotlib.pyplot as plt
from matplotlib.patches import ArrowStyle

import numpy as np

```

```python
t=np.linspace(0,np.pi*2,100)
circle_scale=0.75

fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(1,1,1)

ax.plot(circle_scale*np.cos(t), circle_scale*np.sin(t), color='grey', linewidth=1., zorder=0)
#draw xaxis
style=ArrowStyle("->", head_length=0.3, head_width=10, 
                widthA=1.0, widthB=1.0, lengthA=0.2, 
                lengthB=0.2, angleA=0, angleB=0, scaleA=None, scaleB=None)


ax.annotate("", xy=(1.1, 0.), xytext=(-1.1, 0), arrowprops=dict(arrowstyle='->', linewidth=3,), zorder=1)
ax.annotate("", xy=(0., 1.1), xytext=(0, -1.1), arrowprops=dict(arrowstyle='->', linewidth=3), zorder=1)

z_x_p=np.random.normal(loc=np.pi/2, scale=np.pi/2, size=(10,))
z_x_n=np.random.normal(loc=np.pi/2, scale=np.pi/2, size=(10,))

#healthies
ax.scatter(circle_scale*np.cos(z_x_p), circle_scale*np.sin(z_x_p), s=100, linewidths=2, color='#FF91A4', edgecolors='#E0115F', zorder=2)
#schizophrenia
ax.scatter(circle_scale*np.cos(z_x_n), circle_scale*np.sin(z_x_n), s=100, linewidths=2, color='#73C2FB', edgecolors='#446CCF', zorder=2)

ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.axis('off')

ax.scatter(0.9,0.88, s=100, linewidths=2, color='#FF91A4', edgecolors='#E0115F', zorder=2)
ax.text(1.,0.85,'- HC', fontdict={'fontsize': 12})
ax.scatter(0.9, 1.03, s=100, linewidths=2, color='#73C2FB', edgecolors='#446CCF', zorder=2)
ax.text(1.,1.0,'- SZ', fontdict={'fontsize': 12})


#legend


fig.tight_layout()


fig.show()
```
