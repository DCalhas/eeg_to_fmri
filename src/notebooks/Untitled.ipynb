```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

from matplotlib.animation import FuncAnimation
```

```python
a = np.random.random((10, 50,52))
b = np.random.random((10, 50,52))
```

```python
%matplotlib notebook

n_frames = a.shape[0]

fig, axes = plt.subplots(1,2, figsize=(9,4))
ax1, ax2 = axes

ims1 = []
ims2 = []
for frame in range(n_frames):
    im1 = ax1.imshow(a[frame], cmap='gnuplot2', interpolation='nearest', animated=True)
    im2 = ax2.imshow(b[frame], cmap='gnuplot2', interpolation='nearest', animated=True)
    ims1.append([im1])
    ims2.append([im2])

ani1 = animation.ArtistAnimation(fig, ims1, interval=500, blit=True, repeat_delay=1000)
ani2 = animation.ArtistAnimation(fig, ims2, interval=500, blit=True, repeat_delay=1000)

plt.show()
```
